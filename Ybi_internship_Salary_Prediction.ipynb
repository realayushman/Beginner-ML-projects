```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Salary%20Data.csv")
```

```python
salary.head()
```

```python
salary.info()
```

```python
salary.describe()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=salary)
plt.show()
```

```python
#y=salary,x=exp

y=salary['Salary']
x=salary[['Experience Years']]
```

```python
y.shape,x.shape
```

```python
from sklearn.model_selection import train_test_split as train_n_test
```

```python
x_train,x_test,y_train,y_test=train_n_test(x,y,train_size=0.75,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train
```

```python
#step6: select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#step 8: predict
y_pred=model.predict(x_test)
y_pred
```

```python
x_test
```

```python
y_test
```

```python
#step 8: accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
#94% accurate
```
